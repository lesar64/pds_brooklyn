```python
df2[['passenger_count', 'PULocationID', 'DOLocationID',
    'start_month', 'start_day', 'start_hour', 'start_week', 'weekend', 'weekday',
    'start_location_long', 'start_location_lat', 'end_location_long', 'end_location_lat',
    'DOBorough', 'DOservice_zone', 'PUBorough', 'PUservice_zone']]
```

```python
output = pd.DataFrame()
```

```python
trip_distance_model = yellowcab.model.trip_distance_model()
```

```python
output["p_trip_distance"] = trip_distance_model.predict(df2[
            ['passenger_count', 'PULocationID', 'DOLocationID',
            'start_month', 'start_day', 'start_hour', 'start_week', 'weekend', 'weekday',
            'start_location_long', 'start_location_lat', 'end_location_long', 'end_location_lat',
            'DOBorough', 'DOservice_zone', 'PUBorough', 'PUservice_zone']])
```

```python
print("RMSE:", np.sqrt(metrics.mean_squared_error(df2["trip_distance"], output["p_trip_distance"])))
print("MAE:", metrics.mean_absolute_error(df2["trip_distance"], output["p_trip_distance"]))
```
